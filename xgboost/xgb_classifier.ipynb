```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import f1_score, accuracy_score, roc_auc_score, make_scorer
import sklearn.model_selection as ms
from sklearn.pipeline import make_pipeline
from xgboost import XGBClassifier
```

```python
x_train = pd.read_csv('X_train_cc.csv', index_col=0)
x_valid = pd.read_csv('X_valid_cc.csv', index_col=0)
X = pd.concat([x_train, x_valid])
```

```python
y_train = pd.read_csv('Y_train.csv', index_col=0)
y_valid = pd.read_csv('Y_valid.csv', index_col=0)
y = np.asarray(pd.concat([y_train, y_valid])).flatten()
```

```python
class XGBP(XGBClassifier):
    def predict(self, x):
        return self.predict_proba(x)[:,1]
```

```python
params = {'max_depth': 5, 'learning_rate': 0.03754646253312117, 'n_estimators': 10000, 'min_child_weight': 2, 'colsample_bytree': 0.8442956116897381, 'subsample': 0.683296338928301, 'gamma': 13.048293930079065, 'lambda': 0.6515093911969478, 'alpha': 684.9457534784535}
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size=0.1, stratify=y)
```

```python
clf = XGBP(**params)
```

```python
v = clf.fit(X_train, y_train, eval_set=[(X_test, y_test)], eval_metric='auc', early_stopping_rounds=100)
```
